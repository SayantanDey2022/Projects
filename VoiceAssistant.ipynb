```python
import pyttsx3
import speech_recognition as sr
import webbrowser
import datetime
import pyjokes
import pyaudio
import os
import time
import pywhatkit

def unique():
        recognizer=sr.Recognizer()
        with sr.Microphone() as source:
            print("Tell me.....")
            recognizer.adjust_for_ambient_noise(source)
            audio=recognizer.listen(source)
            try:
                print("Searching.....")
                data=recognizer.recognize_google(audio)
                return data
            except sr.UnknownValueError:
                print("Not Understanding")

def sptext():
    while True:
        recognizer=sr.Recognizer()
        with sr.Microphone() as source:
            print("Listening...")
            recognizer.adjust_for_ambient_noise(source)
            audio=recognizer.listen(source)
            try:
                print("Recognizing...")
                data=recognizer.recognize_google(audio)
                return data
            except sr.UnknownValueError:
                print("Not Understanding")
            
def speechtx(x):
    engine=pyttsx3.init()
    voices=engine.getProperty('voices')
    engine.setProperty('voice',voices[1].id)
    rate=engine.getProperty('rate')
    engine.setProperty('rate',130)
    engine.say(x)
    engine.runAndWait()
    
if __name__ == '__main__':
    
    if "hey jarvis" in sptext().lower():
            speechtx("hello i am jarvis and i am here to listen you")
            while True:
                data1=sptext().lower()
                if "your name" in data1:
                    name="my name is jarvis"
                    speechtx(name)
                elif "old are you" in data1:
                    speechtx("i was born on 10th january 2023")
                elif "time" in data1:
                    time=datetime.datetime.now().strftime("%I%M%p")
                    speechtx(time)
                elif "youtube" in data1:
                    speechtx("oneping youtube")
                    webbrowser.open("https://www.youtube.com/")
                elif "google" in data1:
                    speechtx("oneping google")
                    webbrowser.open("https://www.google.com/")
                elif "facebook" in data1:
                    speechtx("oneping facebook")
                    webbrowser.open("https://www.facebook.com/")
                elif "instagram" in data1:
                    speechtx("oneping instagram")
                    webbrowser.open("https://www.instagram.com/")
                elif "map" in data1:
                    speechtx("oneping googlemaps")
                    webbrowser.open("https://www.google.com/maps")
                elif "amazon" in data1:
                    speechtx("oneping amazon")
                    webbrowser.open("https://www.amazon.com/")
                elif "flipkart" in data1:
                    speechtx("oneping flipkart")
                    webbrowser.open("https://www.flipkart.com/")
                elif "myntra" in data1:
                    speechtx("oneping myntra")
                    webbrowser.open("https://www.myntra.com/")
                elif "spotify" in data1:
                    webbrowser.open("https://open.spotify.com/")
                elif "joke" in data1:
                    joke_1=pyjokes.get_joke(language="en",category="neutral")
                    speechtx(joke_1)
                elif "play song" in data1:
                    speechtx("Tell me the name of the song")
                    command=unique().lower()
                    pywhatkit.playonyt(command)
                elif "search" in data1:
                    speechtx("What do you want to search?")
                    command2=unique().lower()
                    pywhatkit.search(command2)
                    
                elif "exit" in data1:
                    speechtx("good bye")
                    break
                    
                    time.sleep(10)
                    
    else:
        print("Not Listening")
        
    
        
```
